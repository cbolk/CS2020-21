### DNA sequence
Write a program that receives in input a string representing a DNA sequence and computes and outputs the number of times the four bases occur in a DNA sequence, as well as the total size

```python
ADENINE = 'A'
CYTOSINE = 'C'
GUANINE = 'G'
THYMINE = 'T'

sequence = raw_input()


nA = 0
nC = 0
nG = 0
nT = 0
size = len(sequence)
for i in range(0,size):
    if sequence[i] == ADENINE:
        nA = nA + 1
    elif sequence[i] == CYTOSINE:
        nC = nC + 1
    elif sequence[i] == GUANINE:
        nG = nG + 1
    elif sequence[i] == THYMINE:
        nT = nT + 1

print(nA, nC, nG, nT, size)
```

### GC ratio & AT/GC ratio
Write a program that receives in input a portion of a gene sequence and computes and outputs the GC ratio, as well as the AT/GC ratio.

```python
ADENINE = 'A'
CYTOSINE = 'C'
GUANINE = 'G'
THYMINE = 'T'

sequence = raw_input()

nA = sequence.count(ADENINE)
nC = sequence.count(CYTOSINE)
nG = sequence.count(GUANINE)
nT = sequence.count(THYMINE)
size = len(sequence)

#debug print
#print(nA, nC, nG, nT, size)

#GC-content
gc = float(nG+nC)/(nA+nT+nG+nC)

#AT/GC ratio
atgc = float(nA+nT)/(nG+nC)

print("GC-ratio: {0:.2f}%".format(gc*100))
print("AT/GC-ratio: {0:.2f}%".format(atgc*100))
```

### GC AT/GC ratio for the entire gene
Write a program that receives a gene sequence given one line at a time, and computes and outputs GC and AT/GC ratios


```python
ADENINE = 'A'
CYTOSINE = 'C'
GUANINE = 'G'
THYMINE = 'T'

nA = 0
nC = 0
nG = 0
nT = 0
totsize = 0
sequence = raw_input()
size = len(sequence)
while size > 0:
	nA = nA + sequence.count(ADENINE)
	nC = nC + sequence.count(CYTOSINE)
	nG = nG + sequence.count(GUANINE)
	nT = nT + sequence.count(THYMINE)
	totsize = totsize + size
	sequence = input()
	size = len(sequence)

#debug print
print(nA, nC, nG, nT, totsize)

#GC-content
gc = float(nG+nC)/(nA+nT+nG+nC)

#AT/GC ratio
atgc = float(nA+nT)/(nG+nC)

print("GC-ratio: {0:.2f}%".format(gc*100))
print("AT/GC-ratio: {0:.2f}%".format(atgc*100))
```

### Amino-acids
Write a program that receives in input a protein sequence and computes and outputs for each amino-acid the number of occurrences and its frequency.

```python
TELSEQ = "ACDEFGHKILMNPQRSTVYW"

sequence = raw_input()
size = len(sequence)

for elem in TELSEQ:
    nocc = sequence.count(elem)
    freq = float(nocc) / size
    strOut = elem + ": " + str(nocc) + " {0:.2f} %".format(freq*100)
    print(strOut)

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ADENINE = 'A'
CYTOSINE = 'C'
GUANINE = 'G'
THYMINE = 'T'

nA = 0
nC = 0
nG = 0
nT = 0
totsize = 0
sequence = input()
size = len(sequence)
elementlist = []
while size > 0:
    elementlist.extend(list(sequence))
    nA = nA + sequence.count(ADENINE)
    nC = nC + sequence.count(CYTOSINE)
    nG = nG + sequence.count(GUANINE)
    nT = nT + sequence.count(THYMINE)
    totsize = totsize + size
    sequence = input()
    size = len(sequence)

df = pd.DataFrame(elementlist)
df.columns = ['BASE']

## visualization only
df = pd.DataFrame(elementlist,columns=['BASE'])
myfmcolor = sns.color_palette(['#4695d6','#fa6e57','#7ebc59','#fed95c'])
sns.set_palette(myfmcolor)
sns.countplot(x='BASE', data=df)
plt.title('Sequence DNA bases')
plt.show()
```

### Neuron statistics
Write a program that acquires the length of a sequence of neurons, concluded when the user inputs 0. It computes and outputs minimal and maximal lengths, average and standard deviation.

```python
neurons = []
val = float(raw_input())
while val != 0:
	neurons.append(val)
	val = float(raw_input())

#debug information    
#print(neurons)

numneurons = len(neurons)
minlength = min(neurons)
maxlength = max(neurons)
avglength = float(sum(neurons)) / numneurons

tot = 0.0
for n in neurons:
	tot = tot + (n-avglength)**2

stddev = (tot/numneurons)**0.5
print(str(minlength), str(maxlength), str(avglength), str(stddev))
```

### Quantiles
Extend the previous program by computing quantiles.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

neurons = []
val = float(raw_input())
while val != 0:
	neurons.append(val)
	val = float(raw_input())

print(neurons)

numneurons = len(neurons)
minlength = min(neurons)
maxlength = max(neurons)
avglength = float(sum(neurons)) / numneurons


neurons.sort()

middle = numneurons / 2		# numneurons // 2  in python 3
th25 = numneurons / 4		# ...
th75 = numneurons * 3 / 4	# ...
q2 = neurons[middle]
q1 = neurons[th25]
q3 = neurons[th75]

print(str(minlength), str(maxlength), str(avglength))
print(str(q1), str(q2), str(q3))
## visualization only
df = pd.DataFrame(neurons,columns=['Length'])
myfmcolor = sns.color_palette(['#4695d6','#fa6e57','#7ebc59','#fed95c'])
sns.set_palette(myfmcolor)
sns.boxplot(y='Length', data=df)
sns.swarmplot(y="Length", data=df, color=".25")
plt.show()
```
