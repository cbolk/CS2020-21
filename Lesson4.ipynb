### DNA sequence
Write a program that receives in input a string representing a DNA sequence and computes and outputs the number of times the four bases occur in a DNA sequence, as well as the total size

```python
ADENINE = 'A'
CYTOSINE = 'C'
GUANINE = 'G'
THYMINE = 'T'

sequence = input()


nA = 0
nC = 0
nG = 0
nT = 0
size = len(sequence)
for i in range(0,size):
    if sequence[i] == ADENINE:
        nA = nA + 1
    elif sequence[i] == CYTOSINE:
        nC = nC + 1
    elif sequence[i] == GUANINE:
        nG = nG + 1
    elif sequence[i] == THYMINE:
        nT = nT + 1

print(nA, nC, nG, nT, size)
```

### GC ratio & AT/GC ratio
Write a program that receives in input a portion of a gene sequence and computes and outputs the GC ratio, as well as the AT/GC ratio.

```python
ADENINE = 'A'
CYTOSINE = 'C'
GUANINE = 'G'
THYMINE = 'T'

sequence = input()

nA = sequence.count(ADENINE)
nC = sequence.count(CYTOSINE)
nG = sequence.count(GUANINE)
nT = sequence.count(THYMINE)
size = len(sequence)

#debug print
#print(nA, nC, nG, nT, size)

#GC-content
gc = float(nG+nC)/(nA+nT+nG+nC)

#AT/GC ratio
atgc = float(nA+nT)/(nG+nC)

print("GC-ratio: {0:.2f}%".format(gc*100))
print("AT/GC-ratio: {0:.2f}%".format(atgc*100))
```

### GC AT/GC ratio for the entire gene
Write a program that receives a gene description given one line at a time, and computes and outputs GC and AT/GC ratios


```python
ADENINE = 'A'
CYTOSINE = 'C'
GUANINE = 'G'
THYMINE = 'T'

nA = 0
nC = 0
nG = 0
nT = 0
totsize = 0
sequence = input()
size = len(sequence)
while size > 0:
	nA = nA + sequence.count(ADENINE)
	nC = nC + sequence.count(CYTOSINE)
	nG = nG + sequence.count(GUANINE)
	nT = nT + sequence.count(THYMINE)
	totsize = totsize + size
	sequence = input()
	size = len(sequence)

#debug print
print(nA, nC, nG, nT, totsize)

#GC-content
gc = float(nG+nC)/(nA+nT+nG+nC)

#AT/GC ratio
atgc = float(nA+nT)/(nG+nC)

print("GC-ratio: {0:.2f}%".format(gc*100))
print("AT/GC-ratio: {0:.2f}%".format(atgc*100))
```

### Telemerase
Write a program that receives in input a telemerase sequence and computes and outputs for each element the number of occurrences and its frequency.

```python
TELSEQ = "ACDEFGHKILMNPQRSTVYW"

sequence = input()
size = len(sequence)

for elem in TELSEQ:
	nocc = sequence.count(elem)
	freq = float(nocc) / size
	strOut = elem + ": " + str(nocc) + " {0:.2f} %".format(freq*100)
	print(strOut)

```
